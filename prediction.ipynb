```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder

import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
##Load the trained model, onehot encoder, standard scaller and Lable encorder files 
model = load_model('model.h5')
#label_encoder_gender.pkl
#onehot_encoder_geo.pkl
#scaler.pkl

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encoder_geo = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

```

```python
#Example Input data
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_data_df = pd.DataFrame([input_data])
input_data_df
```

```python
input_data_df['Gender'] = label_encoder_gender.transform(input_data_df['Gender'])
#label_encoder_gender
geo_encoded_data = onehot_encoder_geo.transform([input_data_df['Geography']])
#onehot_encoder_geo

geo_encoded_data
```

```python
onehot_encoder_geo.get_feature_names_out(["Geography"])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoded_data, columns = onehot_encoder_geo.get_feature_names_out(["Geography"])).astype(int)
geo_encoded_df.head()
```

```python
input_data_df = pd.concat([input_data_df.drop('Geography', axis=1), geo_encoded_df], axis = 1)
input_data_df
```

```python
X_pedict = scaler.transform(input_data_df)
X_pedict
```

```python
prediction = model.predict(X_pedict)
prediction[0][0]
```

```python
if prediction[0][0] < 0.5:
    print("Employee will leave organization")
else:
    print("Employee Continue in organization")
```
